```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('../data/raw/covid19-global-forecasting-week-2/train.csv', parse_dates=['Date'])
df.head()
```

```python
df_weather = pd.read_csv('../data/raw/weather/training_data_with_weather_info_week_2.csv', parse_dates=['Date'])
df_weather = df_weather.drop(['Id', 'Id.1', 'country+province'], axis=1)
df_weather.head()
```

```python
df_weather_us = df_weather[df_weather['Country_Region'] == 'US']
df_weather_us = df_weather_us[df_weather_us['ConfirmedCases'] > 0]
df_weather_us = df_weather_us.drop(['Country_Region'], axis=1)
df_weather_us = df_weather_us.sort_values(['Date', 'Province_State'])
df_weather_us['CasesLog'] = np.log(df_weather_us['ConfirmedCases'])
# df_weather_us['Province_State'].unique()
df_weather_us.info()
```

```python
df_grp = df_weather_us.groupby(['Date']).sum()
# df_grp = df_grp[df_grp['ConfirmedCases'] > 0]

sns.lineplot(x=df_grp.index, y='ConfirmedCases', data=df_grp)
# sns.lineplot(x='Date', y='ConfirmedCases', hue='Province_State', data=df_weather_us)
```

```python
df_grp['CasesLog10'] = np.log10(df_grp['ConfirmedCases'])
df_grp['CasesLN'] = np.log(df_grp['ConfirmedCases'])
```

```python
sns.lineplot(x=df_grp.index, y='CasesLog10', data=df_grp)
```

```python
sns.lineplot(x=df_grp.index, y='CasesLN', data=df_grp)
```

```python
states = pd.get_dummies(df_weather_us['Province_State'])
states.head()
```

```python
df_us = pd.concat([df_weather_us, states], axis=1)
# df_us = df_us.drop(['Province_State'], axis=1)
df_us.info()
```

```python
train = df_us[df_us['Date'] <= '2020-03-18']
y_train_confirmed = train['CasesLog']
y_train_fatal = train['Fatalities']
X_train = train.drop(['ConfirmedCases', 'Fatalities', 'Date', 'CasesLog', 'Province_State'], axis=1)

test = df_us[df_us['Date'] > '2020-03-18']
y_test_confirmed = test['ConfirmedCases']
y_test_fatal = test['Fatalities']
X_test = test.drop(['ConfirmedCases', 'Fatalities', 'Date', 'CasesLog', 'Province_State'], axis=1)

X_train.head()
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train_confirmed)
```

```python
lm.coef_
```

```python
preds = lm.predict(X_test)
preds = np.exp(preds)
```

```python
plt.scatter(y_test_confirmed, preds)
```

```python
np.sqrt(mean_squared_error(y_test_confirmed, preds))
```

```python
df_preds = pd.DataFrame(preds, index=test.index, columns=['Preds'])
df_preds
```

```python
df_preds['Confirmed'] = test['ConfirmedCases']
df_preds['Date'] = test['Date']
df_preds['State'] = test['Province_State']
# df_preds = df_preds.set_index('Date')
df_preds
```

```python
df_grp = df_preds.groupby(['Date']).sum()
# df_grp = df_grp[df_grp['ConfirmedCases'] > 0]

plt.plot(df_grp['Confirmed'])
plt.plot(df_grp['Preds'])
```
