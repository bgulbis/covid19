```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as py

py.init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('../data/raw/covid19-global-forecasting-week-1/train.csv', parse_dates=['Date'])
df.head()
```

```python
df.info()
```

```python
df_grp = df.groupby(['Date', 'Country/Region']).sum().reset_index()
df_grp['Date'] = df_grp['Date'].dt.strftime('%m/%d/%Y')
df_grp['Size'] = df_grp['ConfirmedCases'].pow(0.3)
```

```python
df_grp.head()
```

```python
fig = px.scatter_geo(df_grp, locations='Country/Region', locationmode='country names', color='ConfirmedCases', size='Size',
                    hover_name='Country/Region', projection='natural earth', animation_frame='Date')
# fig.update(layout_coloraxis_showscale=False)
py.offline.iplot(fig)
```

```python
df_weather = pd.read_csv('../data/raw/weather/training_data_with_weather_info.csv', parse_dates=['Date'])
df_weather.head()
```

```python
df_weather_grp = df_weather.groupby(['Date', 'Country/Region']).mean().reset_index()
df_weather_grp['Date'] = df_weather_grp['Date'].dt.strftime('%m/%d/%Y')
# df_weather_grp['Size'] = df_weather_grp['ConfirmedCases'].pow(0.3)
```

```python
fig = px.choropleth(df_weather_grp, locations='Country/Region', locationmode='country names', color='temp', 
                    hover_name='Country/Region', projection='natural earth', animation_frame='Date',
                    color_continuous_scale=px.colors.sequential.RdBu_r)
# fig.update(layout_coloraxis_showscale=False)
py.offline.iplot(fig)
```

```python
df_weather_us = df_weather[df_weather['Country/Region'] == 'US'].copy()
```

```python
df_us_grp = df_weather_us.groupby(['Date', 'Province/State']).sum().reset_index()
df_us_grp['Date'] = df_us_grp['Date'].dt.strftime('%m/%d/%Y')
df_us_grp['Size'] = df_us_grp['ConfirmedCases'].pow(0.3)
```

```python
df_us_grp['Province/State'].unique()
```

```python
not_states = ['Diamond Princess', 'Grand Princess', 'Guam', 'Puerto Rico', 'US', 'United States Virgin Islands', 'Virgin Islands']
df_us_grp = df_us_grp[~df_us_grp['Province/State'].isin(not_states)]
```

# Covid Tracking Dataset

data source: www.covidtracking.com

```python
df_state_daily = pd.read_csv('../data/raw/covidtracking/states-daily.csv', parse_dates=['date', 'dateChecked'])
```

```python
df_state_daily = df_state_daily.sort_values(['date', 'state'])
df_state_daily['date_str'] = df_state_daily['date'].dt.strftime('%m/%d/%Y')
df_state_daily.head()
```

```python
fig = px.scatter_geo(df_state_daily, locations='state', locationmode='USA-states', color='positive', size='total',
                    hover_name='state', scope='usa', animation_frame='date_str', 
                    color_continuous_scale=px.colors.sequential.RdBu_r)
# fig.update(layout_coloraxis_showscale=False)
py.offline.iplot(fig)
```

```python
fig = px.choropleth(df_state_daily, locations='state', locationmode='USA-states', color='positive', 
                    hover_name='state', scope='usa', animation_frame='date_str',
                    color_continuous_scale=px.colors.sequential.Greens)
# fig.update(layout_coloraxis_showscale=False)
py.offline.iplot(fig)
```

```python
ny = df_state_daily[df_state_daily['state'] == 'NY']
fig = px.line(ny, x='date', y='total')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=ny.date, y=ny.positive, name='Positive'))
fig.add_trace(go.Scatter(x=ny.date, y=ny.negative, name='Negative'))
fig.add_trace(go.Scatter(x=ny.date, y=ny.pending, name='Pending'))
# fig.add_trace(go.Scatter(x=ny.date, y=ny.hospitalized, name='Hospitalized'))
# fig.add_trace(go.Scatter(x=ny.date, y=ny.total, name='Total'))
fig.show()
```
